### Example code for a linear regression model

#### Reference:  
Example code for a linear regression model by ischmahl from the repo [machine-learning-OER-Basics](https://github.com/Machine-Learning-OER-Collection/Machine-Learning-OER-Basics) is licensed under [CC-BY 4.0](https://creativecommons.org/licenses/by/4.0/).

#### Used libraries in the code:
* [pandas](https://pandas.pydata.org/docs/index.html#) for data analysing
* [scikit-learn](https://scikit-learn.org/stable/index.html) for machine learning
* [matplotlib](https://matplotlib.org/) for data visualization

#### Steps:
1. Generate dataset.
2. Generate linear regression object.
3. Train and test the data.
4. Evaluate model.

#### First step - Generate dataset

The dataset is generated with [sklearn.datasets.make_regression](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html?highlight=make_regression#sklearn.datasets.make_regression) with one feature (n_features=1).

```python
from sklearn.datasets import make_regression
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=1)
print(X_reg.shape)
print(y_reg.shape)
```

Plot dataset with [matplotlib.pyplot.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

```python
import matplotlib.pyplot as plt
plt.title("generated dataset")
plt.xlabel("input values")
plt.ylabel("output values")
plt.plot(X_reg, y_reg, ".")
```

#### Second step - Generate linear regression object

Use [sklearn.linear_model.LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html).

```python
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
```

#### Third step - Train and test the data

Split the data in training and test data.
Find a linear regression model for the training data by using the function fit().

```python
from sklearn.model_selection import train_test_split
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg)
linear_regression.fit(X_reg_train, y_reg_train)
print(f"slope: {linear_regression.coef_}")
print(f"intercept: {linear_regression.intercept_}")
```

Plot the training data and the linear regression model.

```python
plt.plot(X_reg_train, y_reg_train, ".")
plt.plot(X_reg_train, linear_regression.predict(X_reg_train), linestyle='-')
plt.legend(['training data', 'linear regression model'])
plt.title("linear regression model for training data")
plt.xlabel("input values")
plt.ylabel("output values")
```

Test the model.

```python
plt.plot(X_reg_test, y_reg_test, ".")
plt.plot(X_reg_test, linear_regression.predict(X_reg_test), linestyle='-')
plt.legend(['test data', 'linear regression model'])
plt.title("linear regression model for test data")
plt.xlabel("input values")
plt.ylabel("output values")
```

#### Fourth step - Evaluate the model

Use [R squared](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html).

```python
from sklearn.metrics import r2_score
print(f"R squared: {r2_score(linear_regression.predict(X_reg_test), y_reg_test)}")
```

The value for R squared is very close to one. This indicates a good model fit.
